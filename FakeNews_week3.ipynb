```python
import pandas as pd

df = pd.read_csv("FakeNews_Cleaned.csv", encoding='latin-1')

df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="category", data=df)
plt.title("News Articles per Category")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="label", data=df)
plt.title("Fake vs Real News Distribution")
plt.show()
```

```python
# Create new column for text length
df["text_length"] = df["text"].apply(len)

plt.figure(figsize=(8,5))
sns.histplot(df["text_length"], bins=50, kde=True)
plt.title("Distribution of Article Lengths")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="text_length", y="category", data=df, alpha=0.5)
plt.title("Article Length vs Category")
plt.show()
```

```python
plt.figure(figsize=(10,6))
df["author"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Authors by Number of Articles")
plt.xlabel("Author")
plt.ylabel("Number of Articles")
plt.show()
```
