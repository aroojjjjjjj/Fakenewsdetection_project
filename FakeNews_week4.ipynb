```python
import pandas as pd

df = pd.read_csv("FakeNews_Cleaned.csv", encoding="latin-1")
df.head()
```

```python

df["title_length"] = df["title"].apply(len)
df["text_length"] = df["text"].apply(len)


df["label_num"] = df["label"].apply(lambda x: 1 if x=="FAKE" else 0)

df[["title_length", "text_length", "label_num"]].head()
```

```python
print("Mean text length:", df["text_length"].mean())
print("Median text length:", df["text_length"].median())
print("Variance text length:", df["text_length"].var())
```

```python

numeric_df = df[["title_length", "text_length", "label_num"]]


corr = numeric_df.corr()

print("Correlation Matrix:\n", corr)


print("\nCorrelation with label:")
print(corr["label_num"].sort_values(ascending=False))
```
